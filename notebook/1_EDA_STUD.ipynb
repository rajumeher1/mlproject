```python
gender_uniques = df.gender.unique()
race_uniques = df.race_ethnicity.unique()
parental_uniques = df.parental_level_of_education.unique()
lunch_uniques = df.lunch.unique()
test_uniques = df.test_preparation_course.unique()

print(gender_uniques)
print(race_uniques)
print(parental_uniques)
print(lunch_uniques)
print(test_uniques)
```

```python
# define numerical and categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('We have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# Adding columns for total score and average score
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['avg_score'] = df['total_score'] / 3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['avg_score'].count()
writing_full = df[df['writing_score'] == 100]['avg_score'].count()
math_full = df[df['math_score'] == 100]['avg_score'].count()

print(f"Number of student with full marks in reading : {reading_full}")
print(f"Number of student with full marks in writing : {writing_full}")
print(f"Number of student with full marks in math : {math_full}")
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['avg_score'].count()
writing_less_20 = df[df['writing_score'] <= 20]['avg_score'].count()
math_less_20 = df[df['math_score'] <= 20]['avg_score'].count()

print(f"Number of student with less than marks in reading : {reading_less_20}")
print(f"Number of student with less than marks in writing : {writing_less_20}")
print(f"Number of student with less than marks in math : {math_less_20}")
```

```python
# Multivariate Analysis Using Pieplot
plt.rcParams['figure.figsize'] = (20, 12)

plt.subplot(231)
size = df['gender'].value_counts()
labels = gender_uniques
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='%.0f%%')
plt.title('Gender', fontsize=20)
plt.axis('off')

plt.subplot(232)
size = df['race_ethnicity'].value_counts()
labels = race_uniques
color = ['red', 'green', 'blue', 'cyan', 'orange']

plt.pie(size, colors=color, labels=labels, autopct='%.0f%%')
plt.title('Race_Ethinicity', fontsize=20)
plt.axis('off')

plt.subplot(233)
size = df['parental_level_of_education'].value_counts()
labels = parental_uniques
color = ['red', 'green', 'blue', 'cyan', 'orange', 'grey']

plt.pie(size, colors=color, labels=labels, autopct='%.0f%%')
plt.title('Parental Level of Education', fontsize=20)
plt.axis('off')

plt.subplot(234)
size = df['lunch'].value_counts()
labels = lunch_uniques
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='%.0f%%')
plt.title('Lunch', fontsize=20)
plt.axis('off')

plt.subplot(235)
size = df['test_preparation_course'].value_counts()
labels = test_uniques
color = ['red', 'green']

plt.pie(size, colors=color, labels=labels, autopct='%.0f%%')
plt.title('Test Preparation', fontsize=20)
plt.axis('off')

plt.tight_layout()


plt.show()
```

```python
# Univariate Analysis
f,ax = plt.subplots(1,2, figsize=(20,10)) 
sns.countplot(data=df, x=df['gender'], palette='bright', ax=ax[0],saturation=0.80)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=10)


plt.pie(x=df['gender'].value_counts(), labels=gender_uniques, explode=[0,0.1], autopct='%1.1f%%',shadow=True, colors=['#ff4d4d','#ff8000'])

plt.show()
```

```python
# Bivariate Analysis
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(8,3))

X = ['Total Average', 'Math Average']

female_score = [gender_group['avg_score'][0], gender_group['math_score'][0]]
male_score = [gender_group['avg_score'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.02, male_score, 0.04, label = 'Male')
plt.bar(X_axis + 0.02, female_score, 0.04, label = 'Female')

plt.xticks(X_axis, X)
plt.ylabel('Marks')
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```
